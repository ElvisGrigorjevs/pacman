```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon_df = pd.read_csv('/content/Pokemon.csv')
pokemon_df.head(n=3)
```

```python
type_df = pokemon_df['Type 1'].value_counts().reset_index()
type_df
```

```python
type_df.plot(x='Type 1', y='count', kind='bar', color='Purple', edgecolor='black')
plt.title('Pokemona tipu skaits')
plt.show()
```

```python
pokemon_df['Baldurs Gate 3'] = 'nav šeit'
pokemon_df.head()
```

```python
pokemon_df['Attack + Defense'] = pokemon_df['Attack'] + pokemon_df['Defense']
pokemon_df.head()
```

```python
  pokemon_df = pokemon_df.drop('Baldurs Gate 3', axis=1)
  pokemon_df.head()
```

```python
pokemon_df = pokemon_df.sort_values(by='Attack + Defense', ascending=False)
pokemon_df.head()
```

```python
pokemon_df.to_csv('pokemon1.csv')
```

```python
plt.hist(pokemon_df['HP'], color='green', edgecolor='black', bins=25)
plt.xlabel('HP')
plt.ylabel('Pokemona skaits')

plt.axvline(pokemon_df['HP'].mean(), linestyle='dashed', color='blue')
plt.axvline(pokemon_df['HP'].median(), linestyle='dashed', color='turquoise')
plt.axvline(pokemon_df['HP'].min(), linestyle='dashed', color='red')
plt.axvline(pokemon_df['HP'].max(), linestyle='dashed', color='tan')
```

```python
water_df = pokemon_df[pokemon_df['Type 1'] == 'Water']
water_df.head()

plt.scatter(water_df['HP'], water_df['Defense'], color='Green')
plt.xlabel('HP')
plt.ylabel('Defense')

plt.show()
```

```python
correlation = water_df['HP'].corr(water_df['Defense'])
print(correlation)
```

```python
pokemon_numeric_df = pokemon_df.select_dtypes(include=['number'])
correlation_matrix = pokemon_numeric_df.corr()
correlation_matrix
```

```python
heatmap = plt.imshow(correlation_matrix, cmap='coolwarm');

plt.colorbar(heatmap, fraction=0.05, pad=0.04)

plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90);
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns);

for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}", ha='center', va='center', color='black')

plt.title('Pokemon Correlation matrix')

plt.show()
```
